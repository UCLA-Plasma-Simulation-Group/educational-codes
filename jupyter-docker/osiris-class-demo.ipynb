# Running OSIRIS with Jupyter Notebook

This is an overview of how to run OSIRIS and analyze output within a Jupyter Notebook.

## Requirements
* Download and install Docker
* Download the following files (which are available in the GitHub repo)
 * osiris-1D.e
 * osiris-input.txt
 * rundocker
 * osiris.py
* Open a terminal window and execute the command `docker pull benjum/jdock3`  
 * This will setup the docker image on your computer.

## Starting the OSIRIS-Jupyter environment:
1. Open a terminal window
2. Navigate to the desired folder for input/output
 1. This folder must contain the osiris-1D.e and osiris-input.txt files
3. Execute the command: `./rundocker`
 1. You must have permissions to execute the script
 2. Type `chmod 755 rundocker` if necessary
4. Open a web browser and go to `http://localhost:8888`

# Running OSIRIS

In order to run an OSIRIS simulation, you will need an input deck.

The default input deck is "osiris-input.txt".  

You may create new input decks within the notebook:
* Go to the "Home" tab
* Duplicate the "osiris-input.txt" file (or another input deck)
* Click on the file to open up a separate tab for editing the file
* Save when finished
* The file can be renamed within the "Home" tab or within the editing Tab

To run a simulation, execute:
`osiris.runosiris`

```python
dirname = 'osiris-input-test'
osiris.runosiris(rundir=dirname,inputfile='osiris-input-test.txt')
```

`osiris.runosiris(rundir='',inputfile='osiris-input.txt')`:
* rundir
 * name of the directory in which simulation output will be generated
 * default value = base directory
 * if the named directory does not exist, it will be created
 * if the named directory does exist, its data will be overwritten
* inputfile
 * the name of the input file
 * this file should ideally be in the main home directory, but relative paths can also be given

# Analyzing output

The following functions are currently available for analysis:
* **field**
 * Plots field vs space at specific time
 * command: `osiris.field(rundir='', dataset='e1', time=0, xlim=[-1,-1], ylim=[-1,-1], zlim=[-1,-1], plotdata=[])`
* **phasespace**
 * Plots phasespace at specific time
 * 1D or 2D plot, depending on the phasespace quantity
 * command: `osiris.phasespace(rundir='', dataset='p1x1', species='electrons', time=0, xlim=[-1,-1], ylim=[-1,-1], zlim[-1,-1], plotdata=[])`
* **lineinteract**
 * Plots 1D lineout of field
 * command: `osiris.lineinteract(rundir='', dataset='e1', xlim=[-1,-1], ylim=[-1,-1], zlim=[-1,-1], plotdata=[])`
* **phaseinteract**
 * Plots 1D lineout of field
 * command: `osiris.phaseinteract(rundir='', dataset='p1x1', species='electrons', xlim=[-1,-1], ylim=[-1,-1], zlim=[-1,-1], plotdata=[])`
* **xtplot**
 * Plots 1D lineout of field
 * command: `osiris.xtplot(rundir='', dataset='e3', xlim=[-1,-1], ylim=[-1,-1], zlim=[-1,-1], plotdata=[])`
* **wkplot**
 * Plots 1D lineout of field
 * command: `osiris.wkplot(rundir='', dataset='e3', klim=[-1,-1], wlim=[-1,-1], zlim=[-1,-1], plotdata=[])`
