# Running OSIRIS with Jupyter Notebook

This is an overview of how to run OSIRIS and analyze output within a Jupyter Notebook.

# Inside the Notebook

The following lines must always be executed before running anything else:

```python
import osiris
%matplotlib inline
```

# Running OSIRIS

In order to run an OSIRIS simulation, you will need an input deck.

The default input deck is "osiris-input.txt".  

You may create new input decks within the notebook:
* Go to the "Home" tab
* Duplicate the "osiris-input.txt" file (or another input deck)
* Click on the file to open up a separate tab for editing the file
* Save when finished
* The file can be renamed within the "Home" tab or within the editing Tab

To run a simulation, execute:
`osiris.runosiris`

```python
dirname = 'test'
osiris.runosiris(rundir=dirname,inputfile='osiris-input-test.txt')
```

`osiris.runosiris(rundir='',inputfile='osiris-input.txt')`:
* rundir
 * name of the directory in which simulation output will be generated
 * default value = base directory
 * if the named directory does not exist, it will be created
 * if the named directory does exist, its data will be overwritten
* inputfile
 * the name of the input file
 * this file should ideally be in the main home directory, but relative paths can also be given

# Analyzing output

* **field**
 * Plots field vs space at specific time
 * command: `osiris.field(rundir='', dataset='e1', time=0, xlim=[-1,-1], ylim=[-1,-1], zlim=[-1,-1], plotdata=[])`
* **phasespace**
 * Plots phasespace at specific time
 * 1D or 2D plot, depending on the phasespace quantity
 * command: `osiris.phasespace(rundir='', dataset='p1x1', species='electrons', time=0, xlim=[-1,-1], ylim=[-1,-1], zlim[-1,-1], plotdata=[])`
* **lineinteract**
 * Plots 1D lineout of field with interactive button to scroll through time
 * command: `osiris.lineinteract(rundir='', dataset='e1', xlim=[-1,-1], ylim=[-1,-1], zlim=[-1,-1], plotdata=[])`
* **phaseinteract**
 * Plots phasespace with interactive button to scroll through time
 * command: `osiris.phaseinteract(rundir='', dataset='p1x1', species='electrons', xlim=[-1,-1], ylim=[-1,-1], zlim=[-1,-1], plotdata=[])`
* **plot_xt** 
 * Plots time vs space of field
 * command: `osiris.plot_xt(rundir, TITLE='', b0_mag=0.0, plot_or=3, show_theory=False)`
* **plot_log_xt** 
 * Plots time vs space of field with logarithmic scale for field amplitude
 * command: `osiris.plot_log_xt(PATH, TITLE)`
* **plot_wk**
 * Plots frequency vs wavenumber of field
 * command: `osiris.plot_wk(rundir, TITLE='', vth=0.1, b0_mag=0.0, plot_or=1, show_theory=False, background=0.0, wlim=-1, klim=-1, zlim=[-1,-1], iaw=False)`
* **xtplot**
 * Plots time vs space of field
 * command: `osiris.xtplot(rundir='', dataset='e3', xlim=[-1,-1], ylim=[-1,-1], zlim=[-1,-1], plotdata=[])`
* **wkplot**
 * Plots frequency vs wavenumber of field
 * command: `osiris.wkplot(rundir='', dataset='e3', klim=[-1,-1], wlim=[-1,-1], zlim=[-1,-1], plotdata=[])`

# plot_xt
`osiris.plot(rundir, TITLE, b0_mag, plot_or, show_theory)`
* rundir: can be set to any directory in the main directory.  You may want to set it to 'dirname' as above.
* TITLE: set to any string, this will be the title of your plot.  For example, TITLE='My Title'
* b0_mag: set this to the magnitude of the magnetic field for the data set you are plotting. For a given input file, the magnitude of the magnetic field is specified in the last 3 characters of the title, for example omode-b01 has b0_mag = 0.1.  
* plot_or: the orientation of the plot, i.e. set plot_or = 1 if you want to create an e1 plot, plot_or = 2 for an e2 plot...etc.
* show_theory: set show_theory=True if you want to display the theory curves on your graph, otherwise set to False.

```python
osiris.plot_xt(rundir=dirname)
```

```python
osiris.plot_xt(rundir=dirname,TITLE='',b0_mag=0.0,plot_or=3,show_theory=False)
```
